```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url ='https://en.wikipedia.org/wiki/Cancer'
```

```python
# we are sending request to the url and taking the response
response=requests.get(url)
response
```

```python
response.text
```

```python
# parsing library - html parser and second is lxml
Soup=BeautifulSoup(response.text,"lxml")
```

```python
Soup
```

```python
for Alink in Soup.find_all ('a'):
    print(Alink)
```

```python
#for Blink in Soup.find_all ('li'):
    #print(Blink)
```

```python
for Alink in Soup.find_all ('a'):
    #print(Alink)
    print(Alink.get('href'))
```

```python
for ptag in Soup.find_all('p'):
  print(ptag)
```

```python
for ptag in Soup.find_all('p'):
  #print(ptag)
  # now second loop for b tag
   for btag in ptag.find_all('b'):
        print(btag)
```
